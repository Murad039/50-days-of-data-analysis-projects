```python
import pandas as pd

products = ["Sugar", "Salt", "oil", "diapers", "rice"]
costs = [3123, 2131, 3244, 4234, 4234]
sales = [23344, 423443, 234234, 42344, 42344]


df = pd.DataFrame({"products": products, "costs": costs, "sales": sales})
df
```

```python
df["profit"] = df["sales"] - df["costs"]
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (12, 8))
plt.bar(df['products'], df['profit'])
plt.xlabel("Products")
plt.ylabel("Profit")
plt.title("Products profit", fontsize = 20)
plt.show()
```

```python
df.plot(x='products', y=['costs', 'profit'], kind = 'line', figsize=(12, 8))

        
```
