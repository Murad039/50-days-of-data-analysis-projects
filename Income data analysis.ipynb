```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\\datafile\\income_data.csv")
df.head()
```

```python
df['Gender'].value_counts()
```

```python
ed_index = df[df['Name']== 'Edward'].index
ed_height = df.loc[ed_index, 'Height']
print('Edward height is ', ed_height.iloc[0])
```

```python
df_females = df.query('Gender == "F"')
df_females
```

```python
df_females.reset_index(inplace = True, drop = True)
df_females
```

```python
avg_income = df_females['Income'].mean()
print(f'average income per female is {avg_income} dollars')
```

```python
df.head()
```

```python
plt.figure(figsize = (10, 6))

df.plot.scatter(x = 'Age', y = ['Income'])
plt.title('correlation between age and income', fontsize = 20)
plt.show()
```

```python
males_over_50 = df[(df['Gender'] == 'M') & (df['Age'] > 50)]['Income'].mean()
females_over_50= df[(df['Gender'] == 'F') &(df['Age'] > 50)]['Income'].mean()
```

```python
print(f'Income of males over 50 is {males_over_50}')
print(f'Income of males over 50 is {females_over_50:.2f}')
```

```python
data = {'Gender' : ['Male', 'Female'],
       'Income' : [males_over_50, females_over_50]}
```

```python
bar_df = pd.DataFrame(data)
bar_df
```

```python
bar_df['Income'] = bar_df['Income'].round(2)
bar_df
```

```python
plt.figure(figsize = (10, 6))

plt.bar(bar_df['Gender'], bar_df['Income'], color =['yellow'])
plt.xlabel('Gender')
plt.ylabel('income')
plt.title('Gender vs income')
plt.show()
```
