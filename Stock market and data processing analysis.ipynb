```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:\\datafile\\fusion_stock_data.csv')
df.head()
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.set_index('Date', inplace = True)
df.head()
```

```python
df.plot(y='Price', figsize = (10, 8))
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock price over time', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize = (10, 8))

plt.plot(df.index, df['Price'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock price over time', fontsize = 20)
plt.show()
```

```python
volume_lowest_price = df.loc[df['Price'].idxmin()]['Volume']
```

```python
vol1 = df['Price'].idxmin()
vol_low = df.loc[vol1, 'Volume']

print(f'Volumne on the lowest price {vol_low}')
```

```python
df['return'] = df['Price'].pct_change()
df.head()
```

```python
df.plot(y = 'return', figsize = (10, 6))
plt.xlabel('Date')
plt.ylabel('Return')
plt.title('Daily returns of stock price', fontsize = 20)
plt.show()
```
