```python
df_copy = df.copy()
df_copy['Annual Salary'] = df_copy['Annual Salary'].str.replace('$', '')
df_copy['Annual Salary'] = df_copy['Annual Salary'].str.replace(',', '').astype(int)
```

```python
df_copy.head()
```

```python
df_copy.dtypes
```

```python
df_copy['Goal conversion rate'] = (df_copy['Goals']/ df_copy['Chances Created']) * 100
df_copy.head()
```

```python
df_copy['Goal conversion rate'] = df_copy['Goal conversion rate'].round(2)
df_copy.head()
```

```python
highest_goal_conversion = df_copy['Goal conversion rate'].idxmax()
player = df_copy.loc[highest_goal_conversion, 'Player Name']
conversion = df_copy.loc[highest_goal_conversion, 'Goal conversion rate']
print(f'The player with highest goal conversion rate is {player} amd his conversion rate is {conversion}')
```

```python
df_copy.set_index(['Player Name', 'Club'], inplace = True)
df_copy.head()
```

```python
Lukaku_salary = df_copy.loc[('Romelu Lukaku', 'Chelsea'), 'Annual Salary']
print(f'Annual salary of romelu lukaku is {Lukaku_salary}')
```

```python
df3 = df_copy.unstack(level = ['Player Name', 'Club'])
df3.head()
```

```python
df3.sort_index(inplace = True)
df3
```
