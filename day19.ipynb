```python
import pandas as pd


names = ["joe", "Phil", "ken", "jos"]
miles_run = [120,80,90,85]
times_in_hrs = [40, 22, 34, 43]

df = pd.DataFrame({"Names": names, "Miles run": miles_run, "Time in hours": times_in_hrs})

df
```

```python
names = ["joe", "root", "md"]
age = [32, 43, 23]
gender = ["M", "F", "M"]

new_df = pd.DataFrame({'Names' :names, "Age" : age, "Gender": gender})

df = pd.concat([df, new_df.drop(columns = "Names")], axis = 1)

df
```

```python
import pandas as pd
import matplotlib.pyplot as plt

products = ["sugar", "Salt", "oil", "Diapers"]
costs = [2234, 1233, 1324, 1433]
sales = [4324, 3242, 2342, 4532]

df = pd.DataFrame({"Products": products, "Costs": costs, "Sales": sales})

df.head()
```

```python
df["Profit"] = df["Sales"] - df["Costs"]
df.head()
```

```python
most_profitbale_index_is = df["Profit"].idxmax()
most_profitble_product_id = df.loc[most_profitbale_index_is, "Products"]
most_profitable_product = print(f"Most profitable product is: {most_profitble_product_id}")
                                
```

```python
plt.figure(figsize = (8, 4))
plt.bar(df['Products'], df['Profit'])
plt.xlabel("Product")
plt.ylabel("Profit")
plt.title("Products Profit", fontsize = 22, color = "Green")
plt.show()
```

```python
least_profitable_product_index = df["Profit"].idxmin()
least_profitable_product = df.loc[least_profitable_product_index, "Products"]
print(f"Least profitable product is: {least_profitable_product}")
```

```python
df.plot(x ="Products", y = ["Costs", "Profit"], kind = "line", figsize = (12, 8))
plt.xlabel("Products")
plt.ylabel("Costs and Profit")
plt.title("Cost and Profit analysis", fontsize = 20)
plt.show()
```

```python
avg_cost = df["Costs"].mean()
print("The avg cost is", avg_cost)
```

```python
avg_profit = df["Profit"].mean()
print("The avg profit is ", avg_profit)
```
