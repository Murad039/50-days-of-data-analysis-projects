```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('D:\\datafile\\toy_sales_data.csv')
df.head()
```

```python
df.dtypes
```

```python
df['Total Sales'] = df['Total Sales'].str.replace('$', '').astype(int)
df.dtypes
```

```python
sales_df = df.groupby('Item')['Total Sales'].sum()
sales_df
```

```python
plt.figure(figsize=(12,8))

plt.bar(sales_df.index, sales_df)
plt.title('uahsgdfgasydf')
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize = (12, 8))

plt.pie(sales_df, labels = sales_df.index, autopct='%1.1f%%')
plt.title('Total sales analysis', fontsize = 20)

plt.show()
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1, figsize = (15,10))

ax[1].bar(sales_df.index,sales_df)
ax[0].plot(sales_df.index, sales_df)
plt.show()
```
