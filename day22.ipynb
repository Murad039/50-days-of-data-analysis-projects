```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
names = ["carol", "Kate", "Jane", "Kuda", "Tito", "kuku"]
age = [23, np.nan, 34, 56, np.nan, 44]

df = pd.DataFrame({"Names": names, "Age": age})
df
```

```python
df.isna().sum()
```

```python
df_copy = df.copy()
df_copy = df_copy.dropna()
df_copy
```

```python
df_copy2 = df.copy()
df_copy2 = df_copy2.fillna(df_copy2["Age"].mean(), inplace = False)

df_copy2
```

```python
gender_values = ["F","F","F","M","M","M"]
df_copy2['gender'] = gender_values
df_copy2
```
